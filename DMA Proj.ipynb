```python
mycursor = mydb.cursor()
mycursor.execute("Select area, count(company_name) as Number_of_Companies from company group by area order by number_of_companies desc;")
```

```python
myresult = mycursor.fetchall()
area= []
number=[]
for x in myresult:
    area.append(x[0])
    number.append(x[1])
print(area)
print(number)
df = pd.DataFrame(list(zip(area, number)),
                 columns=['area','number'])
```

```python
sns.barplot(x='area',y='number',data=df)
plt.show
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(data=df, x='area')
plt.xticks(rotation=75)
plt.title('Areas with most bookings')
plt.xlabel('area')
plt.ylabel('bookings')
```
